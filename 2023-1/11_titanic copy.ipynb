```python
merged_df = merged_df.drop('Age',axis=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
X = merged_df.drop('Survived',axis=1)
y = merged_df['Survived']
```

```python
clfs = [DecisionTreeClassifier(),
        RandomForestClassifier(),
        RandomForestClassifier(criterion='entropy',max_depth=8),
        RandomForestClassifier(criterion='log_loss',max_depth=10),
        SVC(),
        KNeighborsClassifier()]
```

```python
from sklearn.model_selection import cross_val_score

classifiers = []
scores = []

for clf in clfs:
    cv_scores = cross_val_score(clf,X,y, cv=10, scoring='f1')
    for s in cv_scores:
        classifiers.append(str(clf))
        scores.append(float(s))
```

```python
results = pd.DataFrame({'classifier':classifiers,'F-score':scores})
```

```python
sns.set(rc={'figure.figsize':(25,8)})
sns.violinplot(data=results,x='classifier',y='F-score')
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'criterion':['gini','entropy','log_loss'],
              'max_depth':[2,3,4,5,6,7,8,9,10],
              }

grid = GridSearchCV(RandomForestClassifier(),param_grid,cv=10,scoring='f1')
```

```python
grid.fit(X,y)
grid.best_params_
```
